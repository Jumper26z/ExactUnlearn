划分shard

```python
import pandas as pd
import numpy as np
import random

data = pd.read_csv(r'./data/BPR_div_train_and_test/train_dataset_negative.csv', sep='\t', names=['user_id', 'item_i', 'item_j'], engine='python')

random.seed(42)
np.random.seed(42)

shuffled_data = data.sample(frac=1).reset_index(drop=True)

num_subsets = 9

num_samples = len(shuffled_data)
subset_size = num_samples // num_subsets
remainder = num_samples % num_subsets

subsets = []

start_index = 0
for i in range(num_subsets):
    end_index = start_index + subset_size
    if i < remainder:
        end_index += 1
    subset = shuffled_data.iloc[start_index:end_index]
    subsets.append(subset)
    start_index = end_index

import os
output_dir = f'./data/BPR_div_shard{num_subsets}_random'
os.makedirs(output_dir, exist_ok=True)

for i, subset in enumerate(subsets):
    subset.to_csv(f'{output_dir}/sub_train{i+1}.csv', sep='\t', header=False, index=False)  # 保存到文件subset_1.csv, subset_2.csv, 等等

print(f"训练集数据已按要求随机划分为{num_subsets}个子数据集并保存为CSV文件。")
```
