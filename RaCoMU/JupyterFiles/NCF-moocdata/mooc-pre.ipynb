将moocdata里面的数据转化为 userid、Itemid、interaction、timestamp

```python
import pandas as pd

df = pd.read_csv('./data/mooc_data/data.csv', encoding='ANSI')

df['time'] = pd.to_datetime(df['time']).astype('int64') // 10**9  # 转换为秒级时间戳

df['interaction'] = 1
selected_columns = ['stu_id', 'course_index','interaction', 'time']
df_result = df[selected_columns]

df_result.to_csv('./data/mooc_data/interaction.csv',sep=',', header=False, index=False)
```

去掉interaction小于5行的用户的数据

```python
import pandas as pd 

df = pd.read_csv('./data/mooc_data/interaction.csv', sep=',', header=None, names=['userid', 'itemid', 'interaction', 'timestamp'])

interaction_counts = df['userid'].value_counts()

valid_users = interaction_counts[interaction_counts >= 5].index
filtered_df = df[df['userid'].isin(valid_users)]

user_mapping = {user_id: idx for idx, user_id in enumerate(filtered_df['userid'].unique())}
filtered_df['userid'] = filtered_df['userid'].map(user_mapping)

filtered_df.to_csv('./data/mooc_data/filtered.csv',sep=',', header=False, index=False)
```

```python
import pandas as pd

df = pd.read_csv('./data/mooc_data/filtered.csv', sep=',', header=None, names=['userid', 'itemid', 'interaction', 'timestamp'])

num_users = df['userid'].nunique()
num_courses = df['itemid'].nunique()

print(f"Number of Users: {num_users}")
print(f"Number of Courses: {num_courses}")
```
