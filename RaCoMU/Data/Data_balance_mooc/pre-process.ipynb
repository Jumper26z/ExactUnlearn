sub_test for NCF&NCF

```python
import pandas as pd
#n=3
#n=5
#n=9
n = 15

test_data = pd.read_csv(f'./data_div_train_and_test/test_data.csv', sep=',', header=None, names=['user_id', 'item_id', 'interaction'])

sub_train = []
for k in range(1, n + 1):
    sub_train.append(pd.read_csv(f'./data_div/div{n}/sub_train{k}.csv', sep=',', header=None, names=['user_id', 'item_id', 'interaction']))

user_to_subset = {}
for i, subset in enumerate(sub_train):
    for user_id in subset['user_id'].unique():
        user_to_subset[user_id] = i

user_to_subset_df = pd.DataFrame(list(user_to_subset.items()), columns=['user_id', 'subset_index'])
test_data_merged = test_data.merge(user_to_subset_df, on='user_id', how='left')

for i in range(n):
    subset_data = test_data_merged[test_data_merged['subset_index'] == i].drop(columns=['subset_index'])
    subset_data.to_csv(f'./data_div/div{n}/sub_test{i + 1}.csv', sep=',', header=False, index=False)
```

随机分配用户，但保持平衡性：

```python
import pandas as pd
import numpy as np

df = pd.read_csv('./data_div_train_and_test/train_data.csv')

np.random.seed(42)

df = df.sample(frac=1).reset_index(drop=True)
blocks = [df.iloc[i:i+5] for i in range(0, len(df), 5)]
n = 15
subsets = [[] for _ in range(n)]
for i, block in enumerate(blocks):
    subsets[i % n].append(block)

for i, subset in enumerate(subsets):
    subset_df = pd.concat(subset)
    subset_df.to_csv(f'./random_div/div{n}/sub_train{i+1}.csv', index=False,header=False)
```
